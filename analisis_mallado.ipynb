```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
A = pd.read_csv('tabla_malla.csv',decimal=',')

data = A.apply(pd.to_numeric, errors='coerce')
data
```

```python
fig, ax = plt.subplots(1,2,figsize=(10, 4))
ax1,ax2 = ax
fig.suptitle('Influencia del tamaÃ±o de malla', fontsize=20)
width = 1/5
ticks = np.arange(1)

ax1.set_ylabel('Velocidad vertical mÃ¡xima [m/s]')

ax1.bar(ticks + width, data.loc[0,'Uymax'], width, align="center", label='Malla gruesa', color="red")
ax1.bar(ticks, data.loc[1,'Uymax'], width, align="center", label='Malla media',color="blue")
ax1.bar(ticks - width, data.loc[2,'Uymax'], width, align="center", label='Malla fina', color="green")

ax1.set_xticks([])
ax1.set_xlabel('Ra = 1700')
ax1.legend(loc='best')
fig.tight_layout()

ticks2 = np.arange(2)

ax2.bar(ticks2 + width, (data.loc[3,'Uymax'], data.loc[6,'Uymax']), width, align="center", label='Malla gruesa', color="red")
ax2.bar(ticks2, (data.loc[4,'Uymax'], data.loc[7,'Uymax']), width, align="center", label='Malla media',color="blue")
ax2.bar(ticks2 - width, (data.loc[5,'Uymax'],data.loc[8,'Uymax']), width, align="center", label='Malla fina', color="green")

plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
ax2.tick_params(axis='both', which='both', length=0)

ax2.set_xticks(ticks2)
ax2.set_xticklabels(('Ra = 1800', 'Ra = 1900'))
ax2.legend(loc='best')
```
